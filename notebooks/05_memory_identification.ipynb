# Memory Mechanism Identification

This notebook identifies the physical transport mechanism represented by the recovered memory function H(t).

Models compared:

1. Exponential memory:
$$H(t)=Ae^{-\beta t}$$

2. Power-law memory:
$$H(t)=A(t+\epsilon)^{-\alpha}$$

3. Hybrid memory:
$$H(t)=A_1e^{-\beta t}+A_2(t+\epsilon)^{-\alpha}$$

Model selection uses AIC and BIC criteria.


```python
import sys
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd().parents[0]
sys.path.append(str(ROOT))
```

## Import memory models and model selection tools


```python
from src.memory_models.exponential import exponential_model
from src.memory_models.power_law import power_law_model
from src.memory_models.hybrid_memory import hybrid_model
from src.memory_models.fit_memory_models import fit_memory_model

from src.model_selection.aic import calculate_aic
from src.model_selection.bic import calculate_bic
```

## Generate example recovered memory kernel


```python
time = np.linspace(0,100,300)

memory = hybrid_model(
    time,
    A1=0.6,
    beta=0.04,
    A2=0.4,
    alpha=0.5
)

plt.figure(figsize=(7,4))
plt.plot(time,memory)
plt.xlabel('Time')
plt.ylabel('H(t)')
plt.title('Recovered Memory Function')
plt.show()
```

## Fit exponential, power-law, and hybrid models


```python
models = {}

for name in ['exponential','power_law','hybrid']:
    models[name] = fit_memory_model(
        time,
        memory,
        model=name
    )

print('Models fitted successfully')
```

## Visual comparison of fitted models


```python
plt.figure(figsize=(8,5))

plt.plot(
    time,
    memory,
    'k',
    linewidth=2,
    label='Recovered H(t)'
)

for name,result in models.items():
    plt.plot(
        time,
        result['fitted_memory'],
        label=name
    )

plt.xlabel('Time')
plt.ylabel('Memory')
plt.legend()
plt.show()
```

## Calculate AIC and BIC


```python
parameter_numbers = {
    'exponential': 2,
    'power_law': 2,
    'hybrid': 4
}

results=[]

for name,result in models.items():

    aic = calculate_aic(
        memory,
        result['fitted_memory'],
        parameter_numbers[name]
    )

    bic = calculate_bic(
        memory,
        result['fitted_memory'],
        parameter_numbers[name]
    )

    results.append([name,aic,bic])

results
```

The model with the lowest AIC and BIC provides the most appropriate physical interpretation of the transport memory.
