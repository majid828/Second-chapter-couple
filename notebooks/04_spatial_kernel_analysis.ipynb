# Spatial Kernel Analysis

This notebook demonstrates spatial transport analysis using the spatial probability kernel:

$$G(x,t)=\frac{C(x,t)}{\int C(x,t)dx}$$

The analysis includes:

1. Spatial probability density
2. Plume centroid
3. Spatial variance
4. Higher-order moments
5. Non-Gaussian transport indicators


```python
import sys
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd().parents[0]
sys.path.append(str(ROOT))
```

## Import spatial analysis modules


```python
from src.spatial_kernel.spatial_probability import spatial_probability_density
from src.spatial_kernel.plume_statistics import plume_centroid, plume_variance, plume_moments
from src.spatial_kernel.non_gaussian import skewness, kurtosis
from src.spatial_kernel.characteristic_function import characteristic_function
```

## Generate synthetic spatial plume


```python
x = np.linspace(-20, 20, 500)

# Example asymmetric plume concentration
C = np.exp(-(x-5)**2/8)

plt.figure(figsize=(7,4))
plt.plot(x, C)
plt.xlabel('Position')
plt.ylabel('Concentration')
plt.title('Spatial Concentration Profile')
plt.show()
```

## Convert concentration into spatial probability kernel


```python
G = spatial_probability_density(
    x,
    C
)

print('Probability integral =', np.trapz(G, x))

plt.figure(figsize=(7,4))
plt.plot(x, G)
plt.xlabel('Position')
plt.ylabel('G(x)')
plt.title('Spatial Probability Kernel')
plt.show()
```

## Calculate plume statistics


```python
centroid = plume_centroid(x, G)
variance = plume_variance(x, G)
moments = plume_moments(x, G, order=4)

print('Centroid:', centroid)
print('Variance:', variance)
print('Moments:', moments)
```

## Non-Gaussian plume characteristics


```python
S = skewness(x, G)
K = kurtosis(x, G)

print('Skewness:', S)
print('Kurtosis:', K)
```

## Characteristic function analysis

The characteristic function is:

$$\phi(k)=\int e^{ikx}G(x)dx$$


```python
k_values = np.linspace(-2, 2, 200)

phi = characteristic_function(
    x,
    G,
    k_values
)

plt.figure(figsize=(7,4))
plt.plot(k_values, np.real(phi), label='Real part')
plt.plot(k_values, np.imag(phi), label='Imaginary part')
plt.xlabel('Wave number k')
plt.ylabel('Characteristic function')
plt.legend()
plt.show()
```
