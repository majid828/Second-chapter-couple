# Memory Kernel Recovery

This notebook demonstrates recovery of the temporal memory function H(t) from a synthetic breakthrough curve (BTC).

Workflow:

1. Load synthetic BTC
2. Construct transport kernel matrix
3. Perform inverse memory recovery
4. Compare recovered and true memory functions


```python
import sys
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().parents[0]
sys.path.append(str(ROOT))
```

## Import framework modules


```python
from src.mobile_kernel.parametric_kernel import mobile_kernel
from src.mobile_kernel.kernel_matrix import build_convolution_matrix
from src.inverse_model.bayesian_inverse import recover_memory_kernel
```

## Load synthetic breakthrough curve


```python
btc_file = ROOT / 'data' / 'synthetic' / 'generated_btc' / 'synthetic_btc.csv'

btc_data = pd.read_csv(btc_file)

time = btc_data['time'].values
btc = btc_data['btc'].values

plt.figure(figsize=(7,4))
plt.plot(time, btc)
plt.xlabel('Time')
plt.ylabel('BTC concentration')
plt.title('Synthetic BTC')
plt.show()
```

## Construct mobile transport operator


```python
g = mobile_kernel(
    time,
    m=2,
    b=0.05
)

K = build_convolution_matrix(g)
```

## Recover memory kernel

The inverse problem estimates H(t) from:

$$BTC = K H$$


```python
result = recover_memory_kernel(
    time,
    btc,
    K,
    lam=1e-3,
    noise_variance=1e-4
)

H_recovered = result['memory']

plt.figure(figsize=(7,4))
plt.plot(time, H_recovered, label='Recovered H(t)')
plt.xlabel('Time')
plt.ylabel('Memory function')
plt.legend()
plt.show()
```

The recovered memory kernel can now be used for uncertainty analysis and physical mechanism identification.
